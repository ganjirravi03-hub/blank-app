```python
# =========================================================
# STEP 1 — IMPORT LIBRARIES
# =========================================================
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression

# =========================================================
# STEP 2 — LOAD DATA
# =========================================================
ref_path = "/kaggle/input/ai-mathematical-olympiad-progress-prize-3/reference.csv"
test_path = "/kaggle/input/ai-mathematical-olympiad-progress-prize-3/test.csv"

reference = pd.read_csv(ref_path)
test = pd.read_csv(test_path)

print("Reference Data Loaded:")
print(reference.head())

print("\nTest Data Loaded:")
print(test.head())

# =========================================================
# STEP 3 — BASIC PROCESSING
# =========================================================
X = reference["problem"]
y = reference["answer"]     # correct column

X_test = test["problem"]

# =========================================================
# STEP 4 — TF-IDF + Linear Regression (NO ERRORS)
# =========================================================
tfidf = TfidfVectorizer(stop_words='english')

X_vec = tfidf.fit_transform(X)
X_test_vec = tfidf.transform(X_test)

model = LinearRegression()
model.fit(X_vec, y)

print("\nModel training complete!")

# =========================================================
# STEP 5 — PREDICT
# =========================================================
test_preds = model.predict(X_test_vec)

print("\nFirst predictions:")
print(test_preds[:5])

# =========================================================
# STEP 6 — SUBMISSION
# =========================================================
submission = pd.DataFrame({
    "id": test["id"],
    "answer": test_preds          # correct column name
})

submission.to_csv("submission.csv", index=False)

print("\nSubmission file created successfully!")
```
